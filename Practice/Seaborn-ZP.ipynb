```python
# Kernel Density Plots

sns.kdeplot(df['sepal_length'])
```

```python
sns.kdeplot(df['sepal_length'], color="red", shade=True)
```

```python
sns.kdeplot(df['sepal_length'], color="green", shade=True)
```

```python
df.columns
```

```python
# Plot KDE for all the columns

list_a = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for col in list_a:
    sns.kdeplot(df[col], shade=True)
```
